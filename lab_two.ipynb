```
# ===============================
# ML LAB 2 - ALL EXPERIMENTS
# ===============================

# Install libraries (if needed)
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# CREATE SAMPLE DATASET
# ===============================
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [25, None, 30, 35, 40],
    "Salary": [50000, 60000, None, 80000, 90000],
    "Department": ["IT", "HR", "IT", None, "Finance"]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)


# ===============================
# EXPERIMENT 1: Missing Values
# ===============================
print("\n--- Experiment 1 ---")
print(df.isnull())
print("\nTotal Missing Values:\n", df.isnull().sum())


# ===============================
# EXPERIMENT 2: Remove Missing Rows
# ===============================
print("\n--- Experiment 2 ---")
print("Before:", df.shape)

df_clean = df.dropna()

print("After:", df_clean.shape)


# ===============================
# EXPERIMENT 3: Fill Missing Values
# ===============================
print("\n--- Experiment 3 ---")

df_fill = df.copy()

df_fill["Age"].fillna(df_fill["Age"].mean(), inplace=True)
df_fill["Salary"].fillna(df_fill["Salary"].median(), inplace=True)
df_fill["Department"].fillna("Unknown", inplace=True)

print(df_fill)


# ===============================
# EXPERIMENT 4: Data Types
# ===============================
print("\n--- Experiment 4 ---")

print("Before:\n", df.dtypes)

df["Age"] = df["Age"].astype("float")

print("\nAfter:\n", df.dtypes)


# ===============================
# EXPERIMENT 5: Rename Columns
# ===============================
print("\n--- Experiment 5 ---")

df_rename = df.rename(columns={"Name": "Employee_Name"})

print(df_rename.head())


# ===============================
# EXPERIMENT 6: Replace Values
# ===============================
print("\n--- Experiment 6 ---")

df_replace = df.replace("IT", "Information Technology")

print(df_replace)


# ===============================
# EXPERIMENT 7: Line + Bar Plot
# ===============================
print("\n--- Experiment 7 ---")

df_fill["Salary"].plot(kind="line")
plt.title("Salary Trend")
plt.show()

df_fill["Department"].value_counts().plot(kind="bar")
plt.title("Department Count")
plt.show()


# ===============================
# EXPERIMENT 8: Scatter Plot
# ===============================
print("\n--- Experiment 8 ---")

plt.scatter(df_fill["Age"], df_fill["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()


# ===============================
# EXPERIMENT 9: Histogram
# ===============================
print("\n--- Experiment 9 ---")

df_fill["Salary"].plot(kind="hist")
plt.title("Salary Distribution")
plt.show()


# ===============================
# EXPERIMENT 10: Boxplot
# ===============================
print("\n--- Experiment 10 ---")

df_fill.boxplot(column="Salary")
plt.title("Salary Boxplot")
plt.show()
```
