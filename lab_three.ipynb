```python
# ===============================
# ML LAB 3 - DESCRIPTIVE STATISTICS
# ===============================

# Install libraries (if needed)
!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# LOAD DATASET
# ===============================
# Upload file manually in Colab OR use path
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("StudentsPerformance.csv")

print("First 5 Rows:")
print(df.head())


# ===============================
# IDENTIFY COLUMN TYPES
# ===============================
print("\nColumn Data Types:")
print(df.dtypes)

numerical_cols = df.select_dtypes(include=['int64', 'float64'])
categorical_cols = df.select_dtypes(include=['object'])

print("\nNumerical Columns:\n", numerical_cols.columns)
print("\nCategorical Columns:\n", categorical_cols.columns)


# ===============================
# EXPERIMENT 1: CENTRAL TENDENCY & DISPERSION
# ===============================
print("\n--- Experiment 1 ---")

print("\nMean:\n", numerical_cols.mean())
print("\nMedian:\n", numerical_cols.median())
print("\nMode:\n", numerical_cols.mode().iloc[0])

print("\nMin:\n", numerical_cols.min())
print("\nMax:\n", numerical_cols.max())
print("\nSum:\n", numerical_cols.sum())
print("\nVariance:\n", numerical_cols.var())
print("\nStandard Deviation:\n", numerical_cols.std())


# ===============================
# EXPERIMENT 2: QUARTILES
# ===============================
print("\n--- Experiment 2 ---")

Q1 = numerical_cols.quantile(0.25)
Q2 = numerical_cols.quantile(0.50)
Q3 = numerical_cols.quantile(0.75)

print("Q1:\n", Q1)
print("\nMedian (Q2):\n", Q2)
print("\nQ3:\n", Q3)


# ===============================
# EXPERIMENT 3: CORRELATION & COVARIANCE
# ===============================
print("\n--- Experiment 3 ---")

correlation = numerical_cols.corr()
covariance = numerical_cols.cov()

print("Correlation Matrix:\n", correlation)
print("\nCovariance Matrix:\n", covariance)

# Heatmap for better visualization
sns.heatmap(correlation, annot=True)
plt.title("Correlation Heatmap")
plt.show()


# ===============================
# EXPERIMENT 4: HISTOGRAM & BOXPLOT
# ===============================
print("\n--- Experiment 4 ---")

# Histograms
numerical_cols.hist(figsize=(10,6))
plt.suptitle("Histograms of Numerical Features")
plt.show()

# Boxplots
for col in numerical_cols.columns:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```
