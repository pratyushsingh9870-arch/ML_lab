```python
# ============================
# 🔹 IMPORT LIBRARIES
# ============================
import nltk
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# ============================
# 🔹 DOWNLOAD NLTK DATA (RUN ONCE)
# ============================
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Added to resolve LookupError

# ============================
# 🔹 SAMPLE TEXT (AUTO)
# ============================
text = "Natural Language Processing (NLP) is amazing! It allows machines to understand human language easily."

print("Original Text:\n", text)

# ============================
# 🔹 LOWERCASE
# ============================
text_lower = text.lower()
print("\nLowercase:\n", text_lower)

# ============================
# 🔹 TOKENIZATION
# ============================
tokens = word_tokenize(text_lower)
print("\nTokens:\n", tokens)

# ============================
# 🔹 REMOVE PUNCTUATION
# ============================
tokens_no_punct = [word for word in tokens if word not in string.punctuation]
print("\nWithout Punctuation:\n", tokens_no_punct)

# ============================
# 🔹 REMOVE STOPWORDS
# ============================
stop_words = set(stopwords.words('english'))
tokens_no_stop = [word for word in tokens_no_punct if word not in stop_words]
print("\nWithout Stopwords:\n", tokens_no_stop)

# ============================
# 🔹 STEMMING
# ============================
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in tokens_no_stop]
print("\nStemmed:\n", stemmed)

# ============================
# 🔹 LEMMATIZATION
# ============================
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in tokens_no_stop]
print("\nLemmatized:\n", lemmatized)

# ============================
# 🔹 FINAL TEXT
# ============================
processed_text = " ".join(lemmatized)
print("\nProcessed Text:\n", processed_text)

# ============================
# 🔹 BAG OF WORDS
# ============================
documents = [text, processed_text]

cv = CountVectorizer()
bow = cv.fit_transform(documents)

print("\nBoW Feature Names:\n", cv.get_feature_names_out())
print("\nBoW Matrix:\n", bow.toarray())

# ============================
# 🔹 TF-IDF
# ============================
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)

print("\nTF-IDF Feature Names:\n", tfidf.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())
```
